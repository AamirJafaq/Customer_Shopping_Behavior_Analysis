Fill the null values with median of corresponding column (Review Rating) based on categories.

```python
def fill_na(group_series):
  med=group_series.median()
  group_series=group_series.fillna(med)
  return group_series
data["Review Rating"]=data.groupby('Category')['Review Rating'].transform(fill_na)
```

```python
pd.isnull(data["Review Rating"]).sum()
```

```python
data[data.duplicated()]
```

Renaming the columns, lower case and replace space with _

```python
data.columns=data.columns.str.lower()
data.columns=data.columns.str.replace(" ","_")
data=data.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
(data["discount_applied"]==data["promo_code_used"]).all()
```

Drop promo_code_used column as it is identical to discount_applied.

```python
data.drop(columns=["promo_code_used"], inplace=True)
```

Create four group based on age.

```python
labels=["Young Adult","Adult", "Mid Aged", "Senior"]
data["age_group"]=pd.qcut(data["age"],q=4, labels=labels)
```

```python
data["frequency_of_purchases"].value_counts()
```

Add another column freq_of_purchases_days, numerical form of frequency_of_purchases column.

```python
freq_table={
    "Every 3 Months": 90,
    "Annually": 365,
    "Quarterly": 90,
    "Monthly": 30,
    "Bi-Weekly": 14,
    "Fortnightly": 14,
    "Weekly": 7
}
data["freq_of_purchases_days"] = data["frequency_of_purchases"].map(freq_table)
```

Save the final clean dataset as csv file.

```python
data.to_csv("CLEANcustomers_behaviors.csv", index=False)
```
