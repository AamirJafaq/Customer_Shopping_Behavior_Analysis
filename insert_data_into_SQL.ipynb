```python
import pandas as pd
import sqlalchemy
```

```python
data=pd.read_csv("customers_behaviors.csv")
```

```python
engine = sqlalchemy.create_engine(
    "postgresql+psycopg2://postgres:root@localhost:5432/customer_shopping_behaviour"
)
```

```python
data.to_sql(
        'customers_behavior',
        engine,
        schema='public',
        if_exists='replace',
        index=False
    )
```
